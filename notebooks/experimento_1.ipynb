```python
#Basic imports
import numpy as np
import pandas as pd

#sklearn imports
from sklearn.decomposition import PCA #Principal Component Analysis
from sklearn.manifold import TSNE #T-Distributed Stochastic Neighbor Embedding
from sklearn.cluster import KMeans #K-Means Clustering
from sklearn.preprocessing import StandardScaler #used for 'Feature Scaling'

#plotly imports
import plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
data_path = "StartUpsESG_0602.csv"
#df é o dataframe original
df = pd.read_csv(data_path)
df
```

```python
X = df.copy()
X.isnull().sum()
```

```python
selected = X[['StartUp', 'Country', 'ESG', 'E', 'S', 'G', 'Raised']]
selected
```

```python
cater = selected[['StartUp', 'Country']]
cater
```

```python
numer = selected[['ESG', 'E', 'S', 'G','Raised']]
numer
```

```python
scaler = StandardScaler()
numer_scaled = pd.DataFrame(scaler.fit_transform(numer))
numer_scaled.columns = ['ESG_Scaled', 'E_Scaled', 'S_Scaled', 'G_Scaled','Raised_Scaled']
numer_scaled
```

```python
#numer.columns = ['ESG_Scaled', 'Raised_Scaled']
selected = pd.concat([cater, numer, numer_scaled], axis=1, join='inner')
selected
```

```python
names = selected[['StartUp']]
names_ids = {}
countries = []

i=0
#salvar nomes das empresas e ids
for name in names['StartUp']:
  names_ids[i] = name
  i += 1

country_names = selected[['Country']]
for c in country_names['Country']:
  countries.append(c)

keys = list(names_ids.keys())
names_ids[1]
```

```python
X = selected[['Country','ESG_Scaled', 'E_Scaled', 'S_Scaled', 'G_Scaled','Raised_Scaled']]
X = pd.get_dummies(X)
X
```

```python
#Initialize our model
kmeans = KMeans(n_clusters=4)
#Fit our model
kmeans.fit(X)
```

```python
#Find which cluster each data-point belongs to
clusters = kmeans.predict(X)
```

```python
#Add the cluster vector to our DataFrame, X
X["Cluster"] = clusters
selected['Cluster'] = clusters
```

#PCA

```python
#plotX is a DataFrame containing 6541 values from X
plotX = pd.DataFrame(np.array(X.sample(6541)))

#Rename plotX's columns since it was briefly converted to an np.array above
plotX.columns = X.columns
```

```python
#PCA with three principal components
pca_3d = PCA(n_components=3)
pca = PCA()
pca.fit(numer_scaled)
pca.explained_variance_ratio_
```

```python
#And this DataFrame contains three principal components that will aid us
#in visualizing our clusters in 3-D
PCs = pd.DataFrame(pca_3d.fit_transform(plotX.drop(["Cluster"], axis=1)))
```

```python
PCs.columns = ["PC1", "PC2", "PC3"]
```

```python
plotX = pd.concat([plotX, PCs], axis=1, join='inner')
plotX["dummy"] = 0
plotX.head()
```

```python
#Note that all of the DataFrames below are sub-DataFrames of 'plotX'.
#This is because we intend to plot the values contained within each of these DataFrames.
plotX.insert(0,"ID",keys,True)
selected.insert(0,"ID",keys,True)

cluster0 = plotX[plotX["Cluster"] == 0]
cluster1 = plotX[plotX["Cluster"] == 1]
cluster2 = plotX[plotX["Cluster"] == 2]
cluster3 = plotX[plotX["Cluster"] == 3]

cluster0['ID'] = cluster0['ID'].astype('int')
cluster1['ID'] = cluster1['ID'].astype('int')
cluster2['ID'] = cluster2['ID'].astype('int')
cluster3['ID'] = cluster3['ID'].astype('int')

#substituir os ids pelos nomes das empresas
cluster0_names = []
for id in cluster0['ID']:
  cluster0_names.append(names_ids[id])

cluster1_names = []
for id in cluster1['ID']:
  cluster1_names.append(names_ids[id])

cluster2_names = []
for id in cluster2['ID']:
  cluster2_names.append(names_ids[id])

cluster3_names = []
for id in cluster3['ID']:
  cluster3_names.append(names_ids[id])

cluster0.insert(0,"StartUp",cluster0_names,True)
cluster1.insert(0,"StartUp",cluster1_names,True)
cluster2.insert(0,"StartUp",cluster2_names,True)
cluster3.insert(0,"StartUp",cluster3_names,True)
```
